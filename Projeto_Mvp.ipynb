```python
years = list(range(1991,2024))
```

```python
url_start = 'https://www.basketball-reference.com/awards/awards_{}.html'
```

```python
import requests

for year in years:
    url = url_start.format(year)
    data = requests.get(url)

    with open("mvp/{}.html".format(year),"w+") as f:
      f.write(data.text)
```

```python
import time

years2 = list(range(2021, 2024))

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/113.0.0.0 Safari/537.36"
    )
}

for year in years2:
    url = url_start.format(year)
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        with open(f"mvp/{year}.html", "w", encoding="utf-8") as f:
            f.write(response.text)
        print(f"Salvo com sucesso: {year}")
    else:
        print(f"Erro ao acessar {year}: status {response.status_code}")

    time.sleep(3)  # espera 3 segundos entre cada requisição
```

```python
dfs = []

for year in years:
  with open ("mvp/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page, 'html.parser')
  soup.find('tr',class_="over_header").decompose()
  mvp_table = soup.find(id = 'mvp')
  mvp = pd.read_html(str(mvp_table))[0]
  mvp['Year'] =  year

  dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
mvps['Year'].value_counts().sort_index()
```

```python
mvps.to_csv('mvps.csv')
```
